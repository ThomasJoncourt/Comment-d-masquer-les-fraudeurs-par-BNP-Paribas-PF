```python
# Création du modèle séquentiel
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(256, activation='relu', input_shape=(5148,)),
  #tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation='relu'),
  #tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
# Compiler le modèle
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Créer une fonction de rappel personnalisée pour calculer l'average_precision_score
class AveragePrecisionScoreCallback(Callback):
    def on_epoch_end(self, epoch, logs=None):
        y_pred = self.model.predict(X_test)
        avg_prec_score = average_precision_score(y_test, y_pred)
        print(f"\nEpoch {epoch + 1}: Average Precision Score: {avg_prec_score:.4f}")

# Ajouter la fonction de rappel à la liste des rappels
callbacks_list = [AveragePrecisionScoreCallback()]
# Entraîner le modèle
model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), callbacks=callbacks_list)
y_pred = model.predict(X_test)
average_precision_score(y_test, y_pred) # 0.14
```
